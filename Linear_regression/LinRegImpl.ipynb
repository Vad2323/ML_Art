```python
import numpy as np
from numpy.linalg import inv
```

```python
class LinReg:
    def __init__ (self):
        self.w = 0
        #add to fit> weights contain bias and X contain row of ones as the first row. same for prediction.
        

    def fit(self, X, y):

        #X_new = np.colum_stack (np.ones(X))
        X = np.column_stack((np.ones(X.shape[0]), X))
        
        #We start with MSE, set derivative of MSE to zero, and find the optimal weight.
        #Xt @x) -1 * Xt @y
        #we have loss function MSE. 
        #We calculate derivative (or gradient) of MSE. We set it to zero and solve that equation for weights.
        #The solution we find are the optimal weights. The solution looks like below.
        # This is 'closed solution' formula.
        
        self.w = inv(X.T @X) @ (X.T@y)

    def predict(self, X):
        X = np.column_stack((np.ones(X.shape[0]), X))
        return X@self.w

```
